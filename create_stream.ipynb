My goal in this script is to generate an output file which contains a JSON-formatted dataset of a large number of Tweets. I used Tweepy, a Python package which interfaces with the Twitter API.

I used [this tutorial](http://adilmoujahid.com/posts/2014/07/twitter-analytics/).

This imports the necessary components of the Tweepy library (for interfacing with the Twitter API) and the sys library (for creating the output file).

```python
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import sys
```

This makes it so that output from the script (i.e. the text of the Tweets and the metadata associated with them) is saved to a file.

```python
sys.stdout = open('output.json', 'w')
```

This saves the information generated by the Twitter API to local variables.

```python
access_token = ""
access_token_secret = ""
consumer_key = ""
consumer_secret = ""
```

This creates the class we will use to generate the stream.

```python
class StdOutListener(StreamListener):

    def on_data(self, data):
        print data
        return True

    def on_error(self, status):
        print status
```

This handles Twitter authetification and the connection to Twitter Streaming API.

```python
if __name__ == '__main__':
    l = StdOutListener()
    auth = OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
```

[Twitter API](https://dev.twitter.com/streaming/reference/post/statuses/filter) requires you to filter. I used solution proposed [here](http://stackoverflow.com/questions/27530269/how-can-i-use-tweepy-without-a-filter) in which I filter by location, and set the location to be anywhere on earth (so any geotagged post will be in the stream). Another solution is proposed [here](http://stackoverflow.com/questions/26890605/filter-twitter-feeds-only-by-language) in which you filter to include any Tweet which includes any of the most common words in English.

```python
stream = Stream(auth, l).filter(locations=[-180,-90,180,90])
```
